```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import re
from os import getcwd
path = getcwd()
from ast import literal_eval
from random import choices
df = pd.read_csv(r'C:\Users\noam1\Desktop\Mitochondrial_Gene_comparision_project\Mitochondrial-gene-order-project\DB_csvs\final.csv', index_col = 0)
df.Gene_order = df.Gene_order.apply(literal_eval)
df.Gene_locations = df.Gene_locations.apply(literal_eval)

def order_maker(org):
    """
    Recieve organism name, return its genes ordered by location in the mtDNA
    """
    order = [i.replace('-','').replace('*','') for i in df.loc[org, 'Gene_order']]
    order = [re.sub('\(\w{3}\)','',i) for i in order]
    starts = [i[0] for i in df.loc[org, 'Gene_locations']]
    order = [i for _,i in sorted(zip(starts, order), key = lambda x : x[0])]
    return order

genelist = [i.replace('-','').replace('*','') for i in df.loc['Homo sapiens', 'Gene_order']]
genelist = [re.sub('\(\w{3}\)','',i) for i in genelist]
genelist = list(set(genelist))
```

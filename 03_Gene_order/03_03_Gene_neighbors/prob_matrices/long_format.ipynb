```python
import pandas as pd
import numpy as np
import os, glob

PATH = os.getcwd()
int_ph = ['Chordata','Arthropoda','Echinodermata', 'Mollusca', 'Cnidaria', 'Nematoda']
int_ph = [i  + '_prob.csv' for i in int_ph]
df_list = []
```

```python
c = 0
for i in glob.glob(os.path.join(PATH, '*.csv')):
    name = os.path.split(i)[-1]
    if name in int_ph:
        name = name.replace('_prob.csv', '')
        temp = pd.read_csv(i, index_col = 0)
        temp = temp.where(np.triu(np.ones(temp.shape)).astype(bool))
        temp = temp.reset_index().rename({'index':'gene_1'}, axis = 1)
        temp = temp.melt(id_vars='gene_1', var_name = 'gene_2', value_name = name)
        temp = temp.sort_values(by = 'gene_1')
        temp = temp.dropna()
        if c == 0:
            agg_temp = temp.copy()
        else:
            agg_temp = pd.concat([agg_temp, temp[[name]]], axis = 1)
        c += 1
display(agg_temp.head())
```

```python
agg_temp.to_csv('combined_prob_long.csv', index = False)
```
